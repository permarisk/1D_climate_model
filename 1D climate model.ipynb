### Assumptions
- No Atmosphere
- Incomming short wave radiation from sun ($Q_{\rm SWin} = 340$ [${\rm W} {\rm m}^{-2}$])
- Surface albedo ($\alpha = 0.3$)
- Surface emmit radiation as **blackbody** ($Q_{\rm LWout} = \sigma T_s^4$)
- Stefan-Boltzmann constant ($\sigma = 5.67 \times 10^{-8}$ [${\rm W} {\rm m}^{-2} {\rm K}^{-4}$])


### Energy balance at the surface

$$
\begin{align*}
Q_{\rm SWin} - Q_{\rm SWout} - Q_{\rm LWout} &= 0 \\
(1-\alpha) Q_{\rm SWin} &= \sigma T_s^4 \\
\end{align*}
$$

### Assumptions

- Atmosphere is a single layer of air at temperature $T_a$
- Atmosphere is **completely transparent to shortwave** solar radiation.
- The **surface** absorbs shortwave radiation $(1-\alpha) Q$
- Atmosphere is **partly opaque to infrared** radiation (absorption coefficient = emission coefficient; $\tau = \epsilon = 0.78$)
- Surface emits radiation as **blackbodies** ($\sigma T_s^4$)
- Atmosphere emits radiation as **geybody** ($\sigma \epsilon T_a^4$)
- Atmosphere radiates **equally up and down** ($\sigma \epsilon T_a^4$)
- There are no other heat transfer mechanisms

We can now use the concept of energy balance to ask what the temperature need to be in order to balance the energy budgets at the surface and the atmosphere, i.e. the **radiative equilibrium temperatures**.
